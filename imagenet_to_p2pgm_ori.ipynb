```python
import numpy as np
import struct
from myfuc import *
from PIL import Image
import glob
import cv2
```

```python
base="/home/yy/kamome/Research/nette/imagenette2-320/"
target_base="/home/yy/kamome/Research/nette/p2pgm/"
choice=["val/","train/"]
for mode in choice[:1]:
    tag_list=glob.glob(base+mode+"*")
    for tagPath in tag_list[:1]:
        tagName=tagPath.split("/")[-1]
        img_list=glob.glob(tagPath+"/*")
        target_dir=target_base+mode+tagName+"/"
        os.makedirs(target_dir, exist_ok=True)
        print(target_base+mode+tagName)
        for imgPath in img_list[:1]:
            imgName=imgPath.split("/")[-1][:-5]
            im= cv2.imread(imgPath, cv2.IMREAD_GRAYSCALE)
            ny,nx=im.shape
            content="P2 %d %d 255\n"%(nx,ny)
            for i in range(ny):
                for j in range(nx):
                    content+="%d "%im[i,j]
            content+="\n"
            f=open(target_dir+'%s.pgm'%imgName,'w')
            # f=open("test.pgm",'w')
            f.write(content)
            f.close()
```
